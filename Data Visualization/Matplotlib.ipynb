```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
x = np.linspace(0,5,11)
y = x**2
```

```python
x
```

```python
y
```

```python
plt.plot(x,y)
#plt.show() #for non-jupyter notebook
```

```python
plt.plot(x,y,'r--')
```

```python
plt.xlabel("X label")
plt.ylabel("Y label")
plt.title("Title")
plt.plot(x,y)
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'r')

plt.subplot(1,2,2)
plt.plot(y,x,'b')
```

```python
fig = plt.figure()

axes = fig.add_axes([0.1,0.1,0.8,0.8])
axes.plot(x,y)
axes.set_xlabel("X label")
axes.set_ylabel("Y label")
axes.set_title("Title")
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.1,0.4,0.4,0.4])

axes1.set_title("Bigger")
axes1.plot(x,y)

axes2.set_title("Smaller")
axes2.plot(y,x)
```

```python
fig,axes = plt.subplots(nrows=3,ncols=3)

#axes.plot(x,y)
plt.tight_layout()
```

```python
fig,axes = plt.subplots(nrows=1,ncols=2)

for current_ax in axes:
    current_ax.plot(x,y)
    
axes[0].plot(y,x,"r--")
axes[0].set_title("First plot")
```

```python
fig,axes = plt.subplots(nrows=2,ncols=1,figsize=(8,2))

axes[0].plot(x,y)
axes[1].plot(y,x)
plt.tight_layout()
```

```python
fig.savefig("my_pic.png",dpi=200)
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.set_title("Title")
ax.set_xlabel("X")
ax.set_ylabel("Y")
ax.plot(x,x**2,label="X square")
ax.plot(x,x**3,label="X cube")
ax.legend(loc=(0.1,0.2))
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y,color="blue",lw=10,alpha=0.8,ls="-",marker='*',markersize=20,markerfacecolor="black",markeredgewidth=2,markeredgecolor="red")
ax.set_xlim([0,1])
ax.set_ylim([0,2])
```
